```python
import matplotlib.pyplot as plt
import numpy as np

# Set global font properties
plt.rcParams['font.size'] = 12
plt.rcParams['font.family'] = 'Arial'
plt.rcParams['font.style'] = 'normal'

y = np.array([19.95, 14.46, 12.62, 10.90, 8.12, 6.11, 3.47, 2.95, 2.77, 2.28])

topexporters = ["USA", "France", "Canada", "Australia", "Russia", "Germany", "Kazakhstan", "Ukraine", "Argentina", "UK"]
explode = [0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0]
colors = ['#6B6B6B', '#898989', '#959595', '#A0A0A0', '#C0504D', '#ACACAC', '#B8B8B8', '#C3C3C3', '#CFCFCF', '#DBDBDB']

plt.pie(y, labels = topexporters, explode = explode, wedgeprops = {'linewidth' : 1, 'edgecolor' : 'white' }, colors=colors)
plt.show() 
```
